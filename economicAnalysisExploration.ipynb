```python
import json
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
```

```python
data = pd.read_csv('RPPData.csv')
```

```python
data.head(10)
```

```python
#Store all data on states in a separate df 
states = data.iloc[4:]

#Stores data on US as a whole
federal = data.iloc[:4]
```

```python
#Look to see variation between years for RPP: All Items
df = states[states["Description"] == "RPPs: All items"]

statsAll = {}

for state in df.iterrows():
    #print(state[1][4:])
    rpps = state[1][4:]
    rpps.plot.bar()
    plt.title(state[1][1])
    plt.xlabel("RPPs: All items")
    plt.show()
    statsAll[state[1][1]] = [numpy.mean(rpps), numpy.std(rpps)]
    print("Mean: ", numpy.mean(rpps))
    print("SD: ", numpy.std(rpps))

#There is little to no variation between the years for each state
```

```python
#Look to see variation between years for RPP: Goods
df2 = states[states["Description"] == '  RPPs: Goods']

statsGoods = {}

for state in df2.iterrows():
    #print(state[1][4:])
    rpps = state[1][4:]
    rpps.plot.bar()
    plt.title(state[1][1])
    plt.xlabel("RPPs: Goods")
    plt.show()
    statsGoods[state[1][1]] = [numpy.mean(rpps), numpy.std(rpps)]
    print("Mean: ", numpy.mean(rpps))
    print("SD: ", numpy.std(rpps))

#There is little to no variation between the years for each state
```

```python
#Look to see variation between years for RPP: Services: Rents
df3 = states[states["Description"] == '  RPPs: Services: Rents']

statsRents = {}

for state in df3.iterrows():
    #print(state[1][4:])
    rpps = state[1][4:]
    rpps.plot.bar()
    plt.title(state[1][1])
    plt.xlabel("RPPs: Services: Rents")
    plt.show()
    statsRents[state[1][1]] = [numpy.mean(rpps), numpy.std(rpps)]
    print("Mean: ", numpy.mean(rpps))
    print("SD: ", numpy.std(rpps))

"""
There is little to no variation between the years for each state. But there is quite a bit more variation here
compared to the other descriptions.
"""
```

```python
#Look to see variation between years for RPP: Services: Other
df4 = states[states["Description"] == '  RPPs: Services: Other']

statsOther = {}

for state in df4.iterrows():
    #print(state[1][4:])
    rpps = state[1][4:]
    rpps.plot.bar()
    plt.title(state[1][1])
    plt.xlabel("RPPs: Services: Other")
    plt.show()
    statsOther[state[1][1]] = [numpy.mean(rpps), numpy.std(rpps)]
    print("Mean: ", numpy.mean(rpps))
    print("SD: ", numpy.std(rpps))

#There is little to no variation between the years for each state
```

```python
statsAllDf = pd.DataFrame.from_dict(statsAll, orient='index',columns=['mean', 'std']).reset_index().rename(columns = {'index' : 'state'})
statsGoodsDf = pd.DataFrame.from_dict(statsGoods, orient='index',columns=['mean', 'std']).reset_index().rename(columns = {'index' : 'state'})
statsRentsDf = pd.DataFrame.from_dict(statsRents, orient='index',columns=['mean', 'std']).reset_index().rename(columns = {'index' : 'state'})
statsOtherDf = pd.DataFrame.from_dict(statsOther, orient='index',columns=['mean', 'std']).reset_index().rename(columns = {'index' : 'state'})
```

```python
#Look at different types of RPP per state (RPP is average from 2008 to 2018. Used average because little amount of variation.)
for row in range(len(statsAllDf)):
    xvals = ["All", "Goods", "Rents", "Other"]
    yvals = [statsAllDf.iloc[row]['mean'], statsGoodsDf.iloc[row]['mean'], statsRentsDf.iloc[row]['mean'], statsOtherDf.iloc[row]['mean']]
    plt.bar(xvals, yvals)
    plt.title(statsAllDf.iloc[row]['state'])
    plt.show()
```

```python
# Lets compare the average standard deviations for each type of RPP
# This verifies the observaation from earlier that there's more variation with RPPs: Services: Rents than the other RPP categories

print("RPPs: All items", statsAllDf['std'].mean())
print("RPPs: Goods", statsGoodsDf['std'].mean())
print("RPPs: Services: Rents", statsRentsDf['std'].mean())
print("RPPs: Services: Other", statsOtherDf['std'].mean())
```

```python
#Now Lets compare RPPs by state
xvals = statsAllDf['state'].map(lambda state: state[:2])
yvals = statsAllDf['mean']
plt.bar(xvals,yvals)
plt.title("RPPs: All items")
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()

xvals = statsGoodsDf['state'].map(lambda state: state[:2])
yvals = statsGoodsDf['mean']
plt.bar(xvals,yvals)
plt.title("RPPs: Goods")
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()

xvals = statsRentsDf['state'].map(lambda state: state[:2])
yvals = statsRentsDf['mean']
plt.bar(xvals,yvals)
plt.title("RPPs: Services: Rents")
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()

xvals = statsOtherDf['state'].map(lambda state: state[:2])
yvals = statsOtherDf['mean']
plt.bar(xvals,yvals)
plt.title("RPPs: Services: Other")
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()
```
